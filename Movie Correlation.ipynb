```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read in the data
df = pd.read_csv('D:\Projects\Movie\movies.csv')
```

```python
# Look at the data
df.head()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print(f'{col} - {pct_missing}%')
```

```python
# check data types for columns
df.dtypes
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Cleaning Empty Cells
df.dropna(inplace = True)
```

```python
# change data types

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df.head()
```

```python
# Check sorted "gross" column
df.sort_values(by = ['gross'], inplace = False, ascending = False)
df.head()
```

```python
# Drop any duplicates
df.drop_duplicates()
df.head()
```

```python
# Scatter plot with budget and gross
plt.scatter(x = df['budget'], y = df['gross'])

plt.title('Budget vs Gross')
plt.xlabel('Budget')
plt.ylabel('Gross')

plt.show()
```

```python
# plot budget vs gross using seaborn
sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {'color':'red'}, line_kws = {'color':'blue'})
```

```python
# look correlation
df.corr(numeric_only=True, method = 'pearson') #pearson, kendall, spearman
```

```python
# High correlation between budget and gross

correlation_matrix = df.corr(numeric_only=True, method = 'pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Looks at Company 
df_numeric = df.copy()

for col in df_numeric.columns:
    if df_numeric[col].dtype == 'object':
        df_numeric[col] = df_numeric[col].astype('category')
        df_numeric[col] = df_numeric[col].cat.codes
        

        
```

```python
correlation_matrix = df_numeric.corr(numeric_only=True, method = 'pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matric')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
df_numeric.corr()
```

```python
correlation_mat = df_numeric.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr = high_corr[(high_corr) != 1]

high_corr
```
